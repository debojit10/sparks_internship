```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf

from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```python
data = pd.read_csv(r"C:\Users\Debojit Nath\Downloads\student_scores - student_scores.csv.txt")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
data.plot(x = 'Hours', y = 'Scores', style = 'o')
plt.title("Hours VS Marks")
plt.xlabel("Hours spent")
plt.ylabel("Percentage Scored")
plt.show()
```

```python
hours = np.array(data['Hours'])
hours_normalizer = preprocessing.Normalization(input_shape=[1,])
hours_normalizer.adapt(hours)
```

```python
model = tf.keras.Sequential([
    hours_normalizer,
    layers.Dense(1)
])
model.summary()
```

```python
model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
history = model.fit(
    data['Hours'], data['Scores'],
    epochs=1000,
    validation_split = 0.2)
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()
```

```python
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  plt.ylim([0, 10])
  plt.xlabel('Epoch')
  plt.ylabel('Error [MPG]')
  plt.legend()
  plt.grid(True)
```

```python
plot_loss(history)
```
